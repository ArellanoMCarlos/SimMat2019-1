1. Resolver con sympy la siguientes ecuación diferencial:
$$\frac{dx}{dt}=\frac{x}{t+1}$$

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from scipy.integrate import odeint
sym.init_printing(use_latex='mathjax')
```

```python
# Definimos variables
sym.var('t', real=True, positive=True)
x = sym.Function('x')
# Ecuacion diferencial
ode = sym.Eq(sym.Derivative(x(t),t)-x(t)/(t+1),0)
display(ode)
```

```python
sym.dsolve(ode)
```

2. Obtener la solución numérica.

```python
# Funcion a integrar
def func(x,t):
    return x/(t+1)
```

```python
# Condicion inicial
x0 = 1
# Vector de tiempo
t = np.linspace(0,10,100)

x = odeint(func,x0,t)
```

```python
# Grafica x vs.t
plt.figure(figsize=(8,6))
plt.plot(t, x)
plt.grid()
plt.xlabel('t')
plt.ylabel('x(t)')
plt.show()
```

3. Comparar.

```python
# Grafica x vs.t
plt.figure(figsize=(8,6))
plt.plot(t, x, 'r',label='solucion numerica')
plt.plot(t, t+1, 'b--', label='solucion exacta')
plt.grid()
plt.legend()
plt.xlabel('t')
plt.ylabel('x(t)')
plt.xlim([0, 15])
```
